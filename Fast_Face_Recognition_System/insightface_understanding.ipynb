```python
import numpy as np
import cv2
```

```python
import insightface
```

### 1. Configure Insightface Model

```python
from insightface.app import FaceAnalysis
```

```python
# Load buffalo sc model
app = FaceAnalysis(
  name='buffalo_sc', 
  root='~/.insightface_model/models')

app.prepare(ctx_id=0, det_size=(640,640))
```

### 2. Load Image

```python
img = cv2.imread('test_images/test_image_2.jpg')

# cv2.imshow('image', img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
# cv2.waitKey(1)
```

### 3. Apply image to Insightface model and get results

```python
results = app.get(img)
```

```python
print(results)
```

```python
img_copy = img.copy()
for res in results:
    # bounding box
    x1, y1, x2, y2 = res['bbox'].astype(int)
    cv2.rectangle(img_copy, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # key points
    kps = res['kps'].astype(int)
    for kp in kps:
        cv2.circle(img_copy, kp, 3, (255, 255, 0), -1)

    # detection score
    score = "score: {}%".format(int(res['det_score'] * 100))
    cv2.putText(img_copy, score, (x1, y1), cv2.FONT_HERSHEY_DUPLEX, 0.7, (255, 0, 0), 2)

cv2.imshow('image', img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
